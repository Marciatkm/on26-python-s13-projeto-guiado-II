```python


# Defina os dados da tabela como um dicionário
data = {
    "PersonID": [1, 2, 3, 4, 5],
    "Gender": ["Masculino", "Feminino", "Masculino", "Feminino", "Masculino"],
    "Age": [30, 25, 35, 40, 28]
    #"PersonID": [1, 2 ],
    #"Gender": ["Masculino", "Feminino"],
    #"Age": [30, 25]
}

# Crie um DataFrame a partir do dicionário de dados
df = pd.DataFrame(data)

# Criar um gráfico de barras empilhadas para "Age" por "Gender"
df.set_index("PersonID")[["Age"]].plot(kind="bar", stacked=True, figsize=(10, 6))
plt.title("Idade por Gênero")
plt.xlabel("PersonID")
plt.ylabel("Idade")
plt.show()
```

Esse gráfico de barras para comparar as médias o gráfico resultante mostrará as médias da duração do sono para ambos os gêneros (masculino e feminino) em barras coloridas (azul para masculino e rosa para feminino) com rótulos 

```python
# Leia os dados do CSV
df = pd.read_csv('estilo_vida.csv', sep=',', encoding='latin-1')

# Calcule a média da duração do sono por gênero
sleep_duration = df.groupby('Gender')['Sleep Duration'].mean()

# Crie um gráfico de barras para comparar as médias
sleep_duration.plot(kind='bar', color=['pink', 'blue'])

# Personalize o gráfico
plt.xlabel('Gênero')
plt.ylabel('Média da Duração do Sono')
plt.title('Comparação de Horas de Sono entre  Mulheres (Female) e Homens(Male) ')

# Exiba o gráfico
plt.show()




```
